```python
# By manually 

def varience(data):
    n = len(ages)
    mean = sum(data)/n
    #varience
    deviation = [(x -mean)**2 for x in data]
    varience = sum(deviation)/n
    return varience
```

```python
print(varience(ages))
```

```python
def varience1(data, dof = 0):
    n = len(ages)
    mean = sum(data)/n
    #varience
    deviation = [(x -mean)**2 for x in data]
    varience = sum(deviation)/(n-1)
    return varience
```

```python
varience1(ages)
```

```python
import matplotlib.pyplot as plt 
import scipy.stats as stats
import pylab
def plot_data(sample): 
    plt.figure(figsize=(8,4))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stats.probplot(sample, dist="norm", plot=pylab)
    plt.show()
```

```python
s = np.random.normal(0.5, 0.2, 1000)
plot_data(s)
```

```python
plot_data(df['SepalWidthCm'])
```
